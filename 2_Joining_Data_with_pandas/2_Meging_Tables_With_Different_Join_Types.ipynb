# Other joins

```python
actors_movies = pd.read_csv("data/actors_movies.csv")
print(actors_movies.head())
movie_to_genres = pd.read_pickle("data/movie_to_genres.p")
print(movie_to_genres.head())
scifi_movies = movie_to_genres.loc[movie_to_genres["genre"]=="Science Fiction"]
print(scifi_movies.head())
action_movies = movie_to_genres.loc[movie_to_genres["genre"]=="Action"]
print(action_movies.head())
```

```python
# Merge action_movies to the scifi_movies with right join
action_scifi = action_movies.merge(scifi_movies, on='movie_id', how='right',
                                   suffixes=('_act','_sci'))

# From action_scifi, select only the rows where the genre_act column is null
scifi_only = action_scifi[action_scifi['genre_act'].isnull()]

print(action_scifi)
print(scifi_only)
```

```python
# Merge the movies and scifi_only tables with an inner join
movies_and_scifi_only = movies.merge(scifi_only, how="inner",left_on="id", right_on="movie_id")

# Print the first few rows and shape of movies_and_scifi_only
print(movies_and_scifi_only.head())
print(movies_and_scifi_only.shape)
```

# Merging a table to itself

```python
import pandas as pd
crews = pd.read_pickle("data/crews.p")
```

```python
# Merge the crews table to itself
crews_self_merged = crews.merge(crews, on='id', how='inner',
                                suffixes=('_dir','_crew'))
print(crews_self_merged)
```

```python
# Create a boolean index to select the appropriate rows
boolean_filter = ((crews_self_merged['job_dir'] == 'Director') & 
                  (crews_self_merged['job_crew'] != 'Director'))
direct_crews = crews_self_merged[boolean_filter]
```
